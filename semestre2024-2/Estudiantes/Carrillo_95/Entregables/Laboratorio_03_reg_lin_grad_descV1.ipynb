# Regresión lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$


|Training|Y      | X_1  |
|--------|----------|----------|
|1|$Y^{1}$ | $X_1^{1}$|
|2|$Y^{2}$ | $X_1^{2}$|
|.|.         | .        |
|.|.         | .        |
|.|.         | .        |
|m|$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ parámetros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan más "cercano" a $Y$ para cada $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, así:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la función de coste, que permite obtener los valores más cercanos,  esta minimización podrá ser realizada a través de diferentes métodos el más conocido es el gradiente descendente.

![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)




Supongamos un modelo lineal para realizar la predicción, así nuestro modelo estará basado en la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$


Reemplando el modelo anterior en la función de coste, tenemos que:



\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

# Gradiente Descendente


Para determinar el mínimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un número aleatorios inicial $\omega_i$
- Para descender al mínimo de la función, encontremos un  valor para el cual
el deriva de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el mínimo de la función

# Modelo de *machine learning*: Solución general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

## Gradient descent algorithm  1D

$$
θ_{i+1} = \theta_i - \frac{df}{d \theta}
$$

````python
# Generar datos sintéticos
dataDict = {
    'Y': [0, 1, 2, 3, 4, 5],
    'X_1': [0, 1, 2, 3, 4, 5]
}

data = pd.DataFrame(dataDict)

# Definir la función de hipótesis
def H(X, theta):
    h = X @ theta
    return h

# Definir la función de costo
def costJ(X, theta, Y):
    cost = np.mean((H(X, theta) - Y) ** 2)
    return cost

# Crear los valores de x para theta_1
x = np.linspace(-100, 100, 100)
costValues = np.zeros(len(x))

theta_0 = 0

# Tomar X y Y del dataframe
X = np.c_[np.ones(len(data)), data['X_1']]  # Añadir la columna de unos para theta_0
Y = data['Y'].values.reshape(-1, 1)

# Bucle para calcular los valores de la función de costo
for i in range(len(x)):
    theta_1 = x[i]
    theta = np.array([theta_0, theta_1]).reshape(-1, 1)
    costValues[i] = costJ(X, theta, Y)

# Graficar la función de costo
plt.plot(x, costValues)
plt.xlabel('theta_1')
plt.ylabel('Cost J')
plt.title('Cost Function J(theta_1)')
plt.show()
````

````python
#@title gradient descent


def costDerivative(theta):
  value = 2 * (theta -1)
  return value


def gradientDescent(X, Y, theta, alpha, tol=0.001):
    m = len(Y)
    thetaHist = []

    actualTheta = theta
    thetaHist.append(actualTheta)

    diff = 1
    while tol < diff:
      newTheta = actualTheta - alpha * costDerivative(actualTheta)
      thetaHist.append(newTheta)
      diff = np.abs(newTheta - actualTheta)

      actualTheta = newTheta

    return thetaHist
````

````python
alpha = 0.1
tol = 0.001
theta = np.random.randint(0, 5)

valores = gradientDescent(X, Y, theta, alpha, tol)


plt.plot(valores)
plt.xlabel('Iteration')
plt.ylabel(r'$ \theta_{i+1}$')
plt.title(f'Gradiente descent evolution \n alpha = {alpha}')

plt.hlines(y=1, xmin=0, xmax=len(valores), linestyles='dashed')
plt.grid()
plt.show()

````

### Función de coste teórica en el caso 1D

En el caso unidimensional (1D), tenemos un solo parámetro, \(\theta_1\), que multiplica a la variable de entrada $(x)$. La hipótesis para la regresión lineal en 1D es de la forma:

$$
h_\theta(x) = \theta_0 + \theta_1 x
$$

La función de coste $(J(\theta_0, \theta_1))$ mide el error entre las predicciones del modelo y los valores reales $(y)$ en el conjunto de entrenamiento. Se define como el promedio del error cuadrático medio:

$$
J(\theta_0, \theta_1) = \frac{1}{2m} \sum_{i=1}^{m} \left( h_\theta(x_i) - y_i \right)^2
$$

Donde:
- $( m )$ es el número de ejemplos en el conjunto de entrenamiento.
- $( h_\theta(x_i) = \theta_0 + \theta_1 x_i )$ es la predicción del modelo para el ejemplo (i)-ésimo.
- $( y_i )$ es el valor real para el ejemplo $(i)$-ésimo.

La función de coste en 1D es simplemente la suma de los errores cuadrados de las predicciones frente a los valores reales, dividida entre (2m) para facilitar la derivación del descenso del gradiente.

### Función de coste teórica en el caso 2D

En el caso bidimensional (2D), tenemos dos variables de entrada, $(x_1)$ y $(x_2)$, y la hipótesis para la regresión lineal toma la forma:

$$
h_\theta(x) = \theta_0 + \theta_1 x_1 + \theta_2 x_2
$$

La función de coste se extiende para incluir los dos términos \(x_1\) y \(x_2\), manteniendo la misma estructura de error cuadrático medio:

$$
J(\theta_0, \theta_1, \theta_2) = \frac{1}{2m} \sum_{i=1}^{m} \left( h_\theta(x_i) - y_i \right)^2
$$

Donde:
- $( h_\theta(x_i) = \theta_0 + \theta_1 x_{1i} + \theta_2 x_{2i} )$ es la predicción del modelo para el ejemplo $(i)$-ésimo.
- $( x_{1i} ) y ( x_{2i} )$ son los valores de las variables de entrada $(x_1)$ y $(x_2)$ para el ejemplo $(i)$-ésimo.

### Generalización para (n) dimensiones

En general, para $(n)$ características o variables de entrada, la hipótesis lineal sería:

$$
h_\theta(x) = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \dots + \theta_n x_n
$$

Y la función de coste generalizada sería:

$$
J(\theta_0, \theta_1, \dots, \theta_n) = \frac{1}{2m} \sum_{i=1}^{m} \left( h_\theta(x_i) - y_i \right)^2
$$

Donde cada $(x_{ji})$ es el valor de la característica $(x_j)$ para el ejemplo $(i)$-ésimo, y $(h_\theta(x_i))$ es la predicción del modelo para el ejemplo $(i)$-ésimo.

````python
def gradientDescent(fPrime, xInit, alpha, epsilon, maxIters=10000):
    x = xInit
    for i in range(maxIters):
        grad = fPrime(x)
        xNew = x - alpha * grad
        if abs(xNew - x) < epsilon:
            print(f"Convergió en {i+1} iteraciones.")
            break
        x = xNew
    return x

# Derivada de f(x) = (x - 4)^2 es f'(x) = 2(x - 4)
def fPrime(x):
    return 2 * (x - 4)

# Parámetros
epsilon = 1e-4
xInit = np.random.uniform(0, 10)  # Valor inicial aleatorio
alphas = [0.1, 0.01, 0.001]  # Tres tasas de aprendizaje diferentes

# Probar para diferentes valores de alpha
for alpha in alphas:
    print(f"Alpha: {alpha}")
    xMin = gradientDescent(fPrime, xInit, alpha, epsilon)
    print(f"Valor mínimo encontrado: x = {xMin}\n")
````

````python
# Función F(x, y)
def F(x, y):
    return np.sin(0.5 * x**2 - 0.25 * y**2 + 3) * np.cos(2 * x + 1 - np.exp(y))

# Gradiente de F(x, y)
def gradF(x, y):
    # Derivadas parciales calculadas numéricamente
    h = 1e-5
    dF_dx = (F(x + h, y) - F(x, y)) / h
    dF_dy = (F(x, y + h) - F(x, y)) / h
    return np.array([dF_dx, dF_dy])

# Gradiente descendente modificado para 2 variables
def gradientDescent2D(gradF, xInit, yInit, alpha, epsilon, maxIters=10000):
    x, y = xInit, yInit
    for i in range(maxIters):
        grad = gradF(x, y)
        xNew = x - alpha * grad[0]
        yNew = y - alpha * grad[1]
        if np.sqrt((xNew - x)**2 + (yNew - y)**2) < epsilon:
            print(f"Convergió en {i+1} iteraciones.")
            break
        x, y = xNew, yNew
    return x, y

# Parámetros
epsilon = 1e-4
xInit, yInit = np.random.uniform(-5, 5), np.random.uniform(-5, 5)
alphas = [0.1, 0.01, 0.001]

# Probar para diferentes valores de alpha
for alpha in alphas:
    print(f"Alpha: {alpha}")
    xMin, yMin = gradientDescent2D(gradF, xInit, yInit, alpha, epsilon)
    print(f"Valor mínimo encontrado: x = {xMin}, y = {yMin}\n")

# Gráfica de la función F(x, y) en 3D
x_vals = np.linspace(-5, 5, 400)
y_vals = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = F(X, Y)

fig = plt.figure(figsize=(14, 6))

# Gráfico 3D
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('F(x, y)')
ax.set_title('Gráfico 3D de F(x, y)')

# Mapa de contorno
ax2 = fig.add_subplot(122)
contour = ax2.contourf(X, Y, Z, cmap='viridis')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Mapa de contorno de F(x, y)')
plt.colorbar(contour)

plt.show()
````

````python
# Generar los datos
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2 * X + 0.02 * np.random.random(100)

# Función de coste
def costFunction(theta0, theta1, X, y):
    m = len(y)
    predictions = theta0 + theta1 * X
    return (1 / (2 * m)) * np.sum((predictions - y) ** 2)

# Gradiente de la función de coste
def computeGradient(theta0, theta1, X, y):
    m = len(y)
    predictions = theta0 + theta1 * X
    dtheta0 = (1 / m) * np.sum(predictions - y)
    dtheta1 = (1 / m) * np.sum((predictions - y) * X)
    return dtheta0, dtheta1

# Algoritmo de descenso del gradiente
def gradientDescent(X, y, alpha, epsilon, maxIters=10000):
    theta0, theta1 = np.random.rand(2)  # Inicialización aleatoria de los parámetros
    for i in range(maxIters):
        dtheta0, dtheta1 = computeGradient(theta0, theta1, X, y)
        theta0_new = theta0 - alpha * dtheta0
        theta1_new = theta1 - alpha * dtheta1
        if abs(theta0_new - theta0) < epsilon and abs(theta1_new - theta1) < epsilon:
            print(f"Convergió en {i+1} iteraciones.")
            break
        theta0, theta1 = theta0_new, theta1_new
    return theta0, theta1

# Parámetros para el descenso del gradiente
alpha = 0.1  # Tasa de aprendizaje
epsilon = 1e-4  # Criterio de convergencia

# Aplicar el descenso del gradiente
theta0, theta1 = gradientDescent(X, y, alpha, epsilon)
print(f"Valores óptimos encontrados: θ0 = {theta0}, θ1 = {theta1}")

# Predicción de la regresión lineal
y_pred = theta0 + theta1 * X

# Gráfica de los resultados
plt.scatter(X, y, label='Datos reales')
plt.plot(X, y_pred, color='red', label='Regresión lineal')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Lineal usando Descenso del Gradiente')
plt.legend()
plt.show()
````
