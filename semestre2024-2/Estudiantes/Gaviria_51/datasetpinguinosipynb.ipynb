```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression



# Cargar el dataset
df0 = sns.load_dataset('penguins')
df=pd.DataFrame(df0)

#Aclaración:

# Se usa la denominación de columna para los datos no numéricos y columna caracteristica para los datos numéricos
```

Pregunta 4

```python
# Gráfico de relación entre longitud de pico y masa corporal
plt.figure(figsize=(8, 6))
# Uso de SNS para la creación de las variables de cada columna
sns.scatterplot(data=df, x='flipper_length_mm', y='body_mass_g')
plt.title('Gráfico de Dispersión entre masa y longitud de pico')
plt.xlabel('Longitud de pico (mm)')
plt.ylabel('Masa corporal (g)')
plt.show()

# Calculo de la correlación entre las columnas con el método  .corr

correlation = df['flipper_length_mm'].corr(df['body_mass_g'])
print(f"La correlación entre el sexo y la longitud de aleta en pinguinos Adelie es: {correlation}")
```

Pregunta 13

```python
# Agrupación de las columnas 'island' y 'species' respecto a la columna caracteristica 'bill_length_mm'
# Uso del método .var() para hallar la varianza y .unstack() para un formato mas ancho
varianza_tamaño_poblacional=df.groupby(['island','species'])['bill_length_mm'].var().unstack()
varianza_tamaño_poblacional.plot(kind='bar', figsize=(10, 6))
# Graficación
plt.title('Variación de la Longitud del Pico entre Especies en Distintas Islas')
plt.ylabel('Longitud del Pico (mm)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Isla')
plt.show()
```

Pregunta 17

```python
#Diferencia de masa corporal promedio entre sexos de la especie Chinstrap

# Filtro para la especie Chinstrap
chinstrap_df = df[df['species'] == 'Chinstrap']

# Calculo de la diferencia del promedio de masa corporal entre sexos con el método .mean( )
mean_body_mass_sex = chinstrap_df.groupby('sex')['body_mass_g'].mean()

# Calcular la diferencia restando maximo y minimo hallado con los métodos max() y min()
difference_body_mass = mean_body_mass_sex.max() - mean_body_mass_sex.min()
print(f"La diferencia del promedio de masa corporal entre sexos de la especie Chinstrap es: {difference_body_mass}")
```

Pregunta 18

```python
# Especies con mayor promedio de longitud de aleta

# Agrupación por especie y calculo del promedio de longitud de aleta
mean_flipper_length_species = df.groupby('species')['flipper_length_mm'].mean()

# Hallar la especie con el mayor promedio con el método .idxmax()
species_max_flipper_length = mean_flipper_length_species.idxmax()

print(f"La especie con el mayor promedio de longitud de aleta es: {species_max_flipper_length}")
```

Pregunta 19

```python
#  Especie con longitud de pico mas pequeña en dataset

# Identificación del menor valor de la columna caracteristica bill_length_mm con el método .min()
min_bill_length = df['bill_length_mm'].min()

# Identificación de la especie con el minimo valor
# Se localizan los valores con los métodos .loc[] e .iloc[]
# Primero se localiza la posición con ese valor en el dataframe
# Despues se extrae la información de esa posición
species_min_bill_length = df.loc[df['bill_length_mm'] == min_bill_length, 'species'].iloc[0]

print(f"La longitud de pico más pequeña es: {min_bill_length}")
print(f"La especie con la lonfitud de pico más pequeña es: {species_min_bill_length}")
```
