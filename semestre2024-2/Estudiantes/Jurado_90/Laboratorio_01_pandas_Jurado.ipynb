### **Punto 5**

```python
#5

def PascalCase(frase): #funcion para convertir frase a PascalCase: esta funcion tomara como argumento un string
  lista1=frase.split("_") #Divide el string usando "_" como delimitador, estos elementos quedan en una lista
  lista2=[]               #creamos lista vacía
  for i in lista1:        #recorremos la lista1
    p=i.capitalize()      #a cada elemento de la lista lo hacemos iniciar con mayúscula
    lista2.append(p)      #adicionamos ese elemento que empieza con mayuscula a una nueva lista vacia (lista2)
  r="".join(lista2)       #juntamos los elementos de lista2 con "", es decir, sin ningún separador
  return r                #al haber usado join lo que se retorna es una cadena con los elementos juntos

#otra forma para la función
def PascalCase2(frase):
  return "".join([i.capitalize() for i in frase.split("_")])

for i in df.columns:
  df=df.rename(columns={i:PascalCase(i)})

df
```

```python
#7
X['DiagnosisNumeric'] = X['Target'].map({'B': 0, 'M': 1})

X.head()
```

### **Punto 8**

```python
#8
# Seleccionar solo las columnas numéricas
Xnorm=X.copy()
numeric_cols = Xnorm.select_dtypes(include=['number']).columns

# Normalizar las columnas numéricas
for col in numeric_cols:
  Xnorm[col] = (Xnorm[col] - Xnorm[col].mean()) / Xnorm[col].std()

Xnorm.head()
```

### **Punto 9**

```python
#9
import re

new_columns=['RadiusMean', 'TextureMean', 'PerimeterMean', 'AreaMean', 'SmoothnessMean', 'CompactnessMean', 'ConcavityMean', 'ConcavePointsMean', "SymmetryMean", "FractalDimensionMean"]
caracteristicas=['Radius', 'Texture', 'Perimeter', 'Area', 'Smoothness', 'Compactness', 'Concavity', 'ConcavePoints', "Symmetry", "FractalDimension"]

X_withoutN=X.copy() #quedara el dataframe nombrado sin numeros

for col in X_withoutN.columns:
  columna_similar= re.match(r'^[a-zA-Z_]+', col).group(0) #columna similar seleccionada
  X_withoutN=X_withoutN.rename(columns={col: columna_similar})

series=[]

for i in range(len(caracteristicas)):
  seriei=X_withoutN[caracteristicas[i]].mean(axis=1)
  series.append(seriei)

X_means=pd.concat(series, axis=1)
X_means.columns=new_columns

X_means  #este dataframe contiene la seleccion de las caracteristicas que podian agruparse y además son numéricas

X_means["Target"]=X.Target
X_means["DiagnosisNumeric"]=X.DiagnosisNumeric


X_means
```

```python
X_meansNorm=X_means.copy()
numeric_cols = X_meansNorm.select_dtypes(include=['number']).columns

# Normalizar las columnas numéricas
for col in numeric_cols:
  X_meansNorm[col] = (X_meansNorm[col] - X_meansNorm[col].mean()) / X_meansNorm[col].std()

X_meansNorm
```

### **Punto 10**

```python
#10

# Filter data for each diagnosis
diagnosis_b = X_meansNorm[X_meansNorm['Target'] == 'B']['RadiusMean']
diagnosis_m = X_meansNorm[X_meansNorm['Target'] == 'M']['RadiusMean']

# Create histogram
plt.figure(figsize=(10, 6))  # Adjust figure size if needed

# Plot histograms for each diagnosis with different colors
plt.hist(diagnosis_b, color='orange', alpha=0.7, label='Benigno (B)')
plt.hist(diagnosis_m, color='blue', alpha=0.7, label='Maligno (M)')

# Add labels and title
plt.xlabel('Radio medio')
plt.ylabel('Frequencia')
plt.title('Histograma del radio medio por diagnóstico')
plt.legend()

# Show plot
plt.show()
```

### **Punto 14**

Existen varias opciones de gráficos para entender mejor los datos, aparte de los que se han visto en este notebook están:

1. **Diagramas de dispersión (Scatter plots)**: Relacionan dos variables continuas para ver correlaciones.
2. **Pairplot**: Visualiza relaciones entre varias variables en una matriz de gráficos.
3. **Mapas de calor (Heatmaps)**: Muestran correlaciones entre variables en una matriz visual.
4. **Gráficos de dispersión en 3D**: Relacionan tres variables continuas.
5. **Gráficos de líneas**: Muestran cambios en una variable a lo largo del tiempo.
6. **Gráfico de pastel (Pie chart)**: Muestra proporciones de categorías.
7. **Gráfico de regresión**: Relaciona dos variables y ajusta una línea de tendencia.
