Objetivo: Programar una regresión multivariada


1. Para simular un conjunto de características $x_1$ , $x_2$,..., $x_n$ trabajaremos en la primera parte con dos características de datos aleatorios que presentan un plano y mostraremos que los párametros optimizados se corresponden con el valor esperado.

- Definir la ecuación  $y = 2.1*x_1 - 3.1*x_2$, y generar números aleatorios que pertenecen al plano.

- Realizar un diagrama 3D de los puntos generados aleatoriamente.


Nuestro objetivo será encontrar los valores $\theta_0 = 0, \theta_1=2.1, \theta_1=3.1$ que mejor ajustar el plano, empleando cálculos vectorizados.

```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
import plotly.graph_objects as go
```

```python
m = 100

# Datos aleatorios
x1 = np.random.rand(m)*2 -1
x2 = np.random.rand(m)*2 -1

# Datos del plano
x1s = np.linspace(-1, 1, m)
x2s = np.linspace(-1, 1, m)
X_1s, X_2s = np.meshgrid(x1s, x2s)

# Función
Y = lambda X_1, X_2: 2.1 * X_1 - 3.1 * X_2

fig = go.Figure(go.Surface(x=X_1s, y=X_2s, z=Y(X_1s,X_2s)))
fig.add_trace(go.Scatter3d(x=x1, y=x2, z=Y(x1,x2), mode='markers', marker=dict(size=3, color='white')))
fig.show()
```

2. Inicializar conjunto de parámetros $\Theta$ de manera aleatoria.


```python
theta = np.random.rand(3)
theta
```

3. Construir la matrix X con dimensiones $(n+1, m)$, m es el numero de datos de entrenamiento y (n) el número de caracteristicas.


```python
matrix = np.array([np.ones(m), x1, x2])  # Esta matrix tiene 3 fila (2 variables x) y m datos de entrenamiento
matrix.shape
```

4. Calcular la función de coste(revise cuidosamente las dimensiones de cada matriz):

  - $h = \Theta^{T} X $
  - $\Lambda= (h -Y) $
  - $\Lambda*= (h -Y)^2 $
  - $\Lambda= [\Lambda_1,\Lambda_2, ...,\Lambda_m]$
  - $J = \frac{1}{2m} \sum_{i}^m \Lambda_i $



```python
def H(theta, X):
  return theta.T@X

def costo(H, Y):
  Lambda = (H - Y)
  Lambda2 = (Lambda)**2
  return np.mean(Lambda2), Lambda.shape, Lambda2.shape, Lambda

J = costo(H(theta, matrix), Y(x1, x2))

print('Costo:', J[0])
print('Dimensión de la matriz Lambda:', J[1])
print('Dimensión de la matriz Lambda*:', J[2])
```

5. Aplicar el gradiente descendente:
  - Encontrar el gradiente.
  $\nabla J = \Lambda X.T$
  
  - Actualizar los nuevos parametros:
  $\Theta_{n+1}=\Theta_{n}-\alpha\nabla J$

6. Iterar para encontrar los valores $\Theta$ que se ajustan el plano.

```python
alpha = 0.01      # Taza de aprendizaje
e = 1E-4          # delta x

grad = np.zeros(3)    # generar array con 3 valores para el gradiente

# Iteración
for k in range(100):
  for i, _t in enumerate(theta):    # Función para hallar el gradiente
    Th_new = np.copy(theta)           # par x, y, originales
    H1 = H(Th_new, matrix)
    J = costo(H1, Y(x1, x2))
    grad[i] = J[3]@matrix[i].T      # Calculo del gradiente

  theta = theta - alpha*grad    # Actualizando la lista de parámetros
```

```python
print('Los parametros optimizados son:', theta)

print(f'\nLa nueva ecuación obtenida con este método es: {theta[0]:3f} + {theta[1]:2f}*X1 {theta[2]:2f}*X2')
```
