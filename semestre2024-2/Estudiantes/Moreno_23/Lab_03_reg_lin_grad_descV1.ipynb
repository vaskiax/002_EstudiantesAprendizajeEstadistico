# Regresión lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$


|Training|Y      | X_1  |
|--------|----------|----------|
|1|$Y^{1}$ | $X_1^{1}$|
|2|$Y^{2}$ | $X_1^{2}$|
|.|.         | .        |
|.|.         | .        |
|.|.         | .        |
|m|$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ parámetros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan más "cercano" a $Y$ para cada $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, así:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la función de coste, que permite obtener los valores más cercanos,  esta minimización podrá ser realizada a través de diferentes métodos el más conocido es el gradiente descendente.

![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)




Supongamos un modelo lineal para realizar la predicción, así nuestro modelo estará basado en la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$


Reemplando el modelo anterior en la función de coste, tenemos que:



\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

# Gradiente Descendente


Para determinar el mínimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un número aleatorios inicial $\omega_i$
- Para descender al mínimo de la función, encontremos un  valor para el cual
el deriva de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el mínimo de la función

# Modelo de *machine learning*: Solución general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
from scipy import stats
from sklearn.linear_model import LinearRegression
````

````python
#definicion de x
x=np.linspace(0,10,10)
y=1*x
````

````python
#defino modelo

def H(theta,X):
  h=theta[0]+theta[1]*X

  return h
````

````python
df=pd.DataFrame({'x':x,'y':y})
````

````python
def costo(H,theta0, theta1, X, y):
    # H: Hipótesis del modelo lineal: h = theta0 + theta1 * X
    #theta0: intercepto
    #theta1: pendiente
    #h = theta0 + theta1 * X
    return (1 / (2 * len(y))) * np.sum((H - y) ** 2)

def df0(h,y):
  #Se calcula la derivada con respecto al parametro theta_0
  return (1/len(y))*(h-y)

def df1(h,y,x):
  #Se calcula derivada con respecto al parametro theta_1
  return (1/len(y))*(h-y)*x

````

````python
#3 se asume entonces theta0=0 y se calcula y grafica la función de costo para diferentes
#valores de theta1
theta = [0, 50]  # Valores iniciales de theta
theta0 = []
theta1 = []

alpha = 0.2 # Tasa de aprendizaje
for i in range(100):  # Número máximo de iteraciones
    h = H(theta, x)  # Calcular las predicciones
    #theta[0] -= alpha * df0(h, y).mean()  # Actualizar theta0, en este caso no se actualiza
    theta[1] -= alpha * df1(h, y, x).mean()  # Actualizar theta1

    # Guardar los valores de theta
    theta0.append(theta[0])
    theta1.append(theta[1])

    # Verificar la condición de parada
    if abs(df1(h, y, x).mean()) < 1E-4:
        print("Converge:", theta)
        break

````

````python
plt.plot(df['x'],df['y'],'o',label='Datos')
plt.plot(df['x'],H(theta,df['x']),label='Ajuste lineal dados por el desc.Gradiente')
plt.show()
````

````python
#Se calcula la funcion de coste para diferentes valores de theta1, asumiendo theta0=0
J=np.zeros(len(theta1))
for i in range(len(theta1)):
  J[i]=costo(H([theta[0],theta1[i]],x),theta[0],theta1,x,y)

````

````python
plt.plot(theta1,J,label='funcion de coste')
plt.xlabel('theta1')
plt.ylabel('J')
plt.legend()
plt.show()
````

````python
#4
theta = [30, 50]  # Valores iniciales de theta
theta0 = []
theta1 = []

alpha = 0.2 # Tasa de aprendizaje
for i in range(1000):  # Número máximo de iteraciones
    h = H(theta, x)  # Calcular las predicciones
    theta[0] -= alpha * df0(h, y).mean()  # Actualizar theta0
    theta[1] -= alpha * df1(h, y, x).mean()  # Actualizar theta1

    # Guardar los valores de theta
    theta0.append(theta[0])
    theta1.append(theta[1])

    # Verificar la condición de parada
    if abs(df0(h, y).mean()) < 1E-4 and abs(df1(h, y, x).mean()) < 1E-4:
        print("Converge:", theta)
        break

print("Resultado final de theta:", theta)
````

````python
theta0_vals = np.linspace(min(theta0), max(theta0), 100)  # 100 puntos entre el mínimo y máximo de theta0
theta1_vals = np.linspace(min(theta1), max(theta1), 100)  # 100 puntos entre el mínimo y máximo de theta1

# Crear la cuadrícula de valores de theta0 y theta1
T0, T1 = np.meshgrid(theta0_vals, theta1_vals)

def costo(theta0, theta1, X, y):
    # Hipótesis del modelo lineal: h = theta0 + theta1 * X
    h = theta0 + theta1 * X
    return (1 / (2 * len(y))) * np.sum((h - y) ** 2)

# Crear una matriz  J_vals que contiene los valores de la función de costo para cada punto (theta0, theta1)
J_vals = np.zeros(T0.shape)

for i in range(T0.shape[0]):
  #Este loop est para llenar la matriz (100,100), que contiene todas las combinaciones de theta0, theta1, estos con fines de poder graficar
    for j in range(T0.shape[1]):
        J_vals[i, j] = costo(T0[i, j], T1[i, j], x, y)
````

````python
# Crear la figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
ax.plot_surface(T0, T1, J_vals, cmap='viridis')

# Etiquetas y título
ax.set_xlabel('Theta0')
ax.set_ylabel('Theta1')
ax.set_zlabel('Costo J')
ax.set_title('Superficie de la función de costo J(theta0, theta1)')

plt.show()
````

````python
#8 datos
X2 = np.linspace(0, 1, 100)
Y2 = 0.2 + 0.2*X + 0.02*np.random.random(100)

````

````python
theta = [2, 3]  # Valores iniciales de theta
theta0 = []
theta1 = []

alpha = 0.5 # Tasa de aprendizaje
for i in range(100000):  # Número máximo de iteraciones
    h = H(theta, X2)  # Calcular las predicciones
    theta[0] -= alpha * df0(h, Y2).mean()  # Actualizar theta0
    theta[1] -= alpha * df1(h, Y2, X2).mean()  # Actualizar theta1

    # Guardar los valores de theta
    theta0.append(theta[0])
    theta1.append(theta[1])

    # Verificar la condición de parada
    if abs(df0(h, Y2)[-1]) < 1E-4 and abs(df1(h, Y2, X2)[-1]) < 1E-4:
        print("Converge:", theta)
        break

print("Resultado final de theta:", theta)
````
