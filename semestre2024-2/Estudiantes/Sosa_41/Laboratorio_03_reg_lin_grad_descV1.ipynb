# Regresión lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$


|Training|Y      | X_1  |
|--------|----------|----------|
|1|$Y^{1}$ | $X_1^{1}$|
|2|$Y^{2}$ | $X_1^{2}$|
|.|.         | .        |
|.|.         | .        |
|.|.         | .        |
|m|$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ parámetros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan más "cercano" a $Y$ para cada $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, así:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la función de coste, que permite obtener los valores más cercanos,  esta minimización podrá ser realizada a través de diferentes métodos el más conocido es el gradiente descendente.

![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)




Supongamos un modelo lineal para realizar la predicción, así nuestro modelo estará basado en la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$


Reemplando el modelo anterior en la función de coste, tenemos que:



\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

````python
# 2. Crear una función para calcular la función de coste.

# Función de coste
def cost_func(f, m, x, y, parameters):
    return 1/2*m*np.sum((f(x, parameters) - y)**2)

# Función lineal
def linear_func(x, parameters):
    return parameters[0] + parameters[1]*x
````

# Gradiente Descendente


Para determinar el mínimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un número aleatorios inicial $\omega_i$
- Para descender al mínimo de la función, encontremos un  valor para el cual
el deriva de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el mínimo de la función

# Modelo de *machine learning*: Solución general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

````python
# 6. Construya un algoritmo en el que emplee el gradiente descente para determinar el minimo de una función, determine 
# el mínimo con una error epsilon de  1E-4, pruebe su algoritmo para $f(x)= (x-4)^2$ y al menos 3 valores de $\alpha$

# Algoritmo de gradiente descendente en 1D
def gradient_descent(func, deriv_func, initial_value, learning_rate, epsilon=1e-4, n=1000):
    x = initial_value
    for i in range(n):
        grad = deriv_func(x)
        next_x = x - learning_rate * grad
        if abs(next_x - x) < epsilon:
            return next_x, func(next_x)
        x = next_x
    return x, func(x)  # Retorna el valor final si no se cumple el criterio

# Función
def f(x):
    return (x - 4)**2

# Derivada de la función
def f_prime(x):
    return 2 * (x - 4)

# Parámetros iniciales
initial_value = 0.0
learning_rates = [0.1, 0.01, 0.001]

results = {alpha: gradient_descent(f, f_prime, initial_value, alpha) for alpha in learning_rates}
results
````

````python
# 7. Encontrar el mínimo de la siguiente función a través del metodo del gradiente descendente 
# https://en.wikipedia.org/wiki/Gradient_descent:


# \begin{equation}
# F(x,y) = \sin \left( \frac{1}{2}x^2-\frac{1}{4}y^2 +3\right) \cos (2x+1-e^y)
# \end{equation}

# Realice una gráfica de la función en 3D, y un mapa de contorno de la función.

# F(x, y)
def F(x, y):
    return np.sin(0.5 * x**2 - 0.25 * y**2 + 3) * np.cos(2 * x + 1 - np.exp(y))

# Espacio para la gráfica en 3D y el mapa de contorno
x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = F(X, Y)

# Gráfico en 3D
fig = plt.figure(figsize=(12, 6))

# Subplot 1: Superficie 3D
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k', alpha=0.7)
ax1.set_title("3D Surface Plot of F(x, y)")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("F(x, y)")

# Subplot 2: Mapa de contorno
ax2 = fig.add_subplot(122)
contour = ax2.contourf(X, Y, Z, cmap='viridis', levels=50)
plt.colorbar(contour, ax=ax2)
ax2.set_title("Contour Map of F(x, y)")
ax2.set_xlabel("x")
ax2.set_ylabel("y")

plt.tight_layout()
plt.show()
````

````python
# Determine el valor mínimo de la función con el metodo del gradiente descendente

# Derivadas parciales de F respecto a x e y
def F_partial_x(x, y):
    return np.cos(0.5 * x**2 - 0.25 * y**2 + 3) * np.cos(2 * x + 1 - np.exp(y)) * x - \
           2 * np.sin(0.5 * x**2 - 0.25 * y**2 + 3) * np.sin(2 * x + 1 - np.exp(y))

def F_partial_y(x, y):
    return -0.5 * np.sin(0.5 * x**2 - 0.25 * y**2 + 3) * np.cos(2 * x + 1 - np.exp(y)) * y - \
           np.exp(y) * np.sin(0.5 * x**2 - 0.25 * y**2 + 3) * np.cos(2 * x + 1 - np.exp(y))

# Algoritmo de gradiente descendente para funciones en 2D
def gradient_descent_2d(func, grad_x, grad_y, initial_values, learning_rate, epsilon=1e-4, n=1000):
    x, y = initial_values
    for i in range(n):
        grad_x_val = grad_x(x, y)
        grad_y_val = grad_y(x, y)
        next_x = x - learning_rate * grad_x_val
        next_y = y - learning_rate * grad_y_val
        if np.sqrt((next_x - x)**2 + (next_y - y)**2) < epsilon:
            return (next_x, next_y), func(next_x, next_y)
        x, y = next_x, next_y
    return (x, y), func(x, y)  # Valor final si no se cumple el criterio

# Parámetros iniciales para el gradiente descendente en 2D
initial_values = (1.0, 1.0)
learning_rate = 0.01

# Encontrar el mínimo de la función F usando el gradiente descendente
min_point, min_value = gradient_descent_2d(F, F_partial_x, F_partial_y, initial_values, learning_rate)
min_point, min_value
````

````python
# 8. Construya un algorítmo que permita determinar una regresión lineal.

# Datos de entrada
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2 * X + 0.02 * np.random.random(100)

# Modelo de regresión lineal
def model(X, theta_0, theta_1):
    return theta_0 + theta_1 * X

# Función de coste
def cost_function(X, y, theta_0, theta_1):
    m = len(y)
    predictions = model(X, theta_0, theta_1)
    return (1 / (2 * m)) * np.sum((predictions - y)**2)

# Parámetros iniciales
theta_0_init = 0.0
theta_1_init = 0.0

# Calcular el costo inicial
initial_cost = cost_function(X, y, theta_0_init, theta_1_init)
initial_cost
````

````python
# Gradiente descendente
def gradient_descent(X, y, theta_0, theta_1, learning_rate, epsilon=1e-4, n=1000):
    m = len(y)
    costs = []
    
    for _ in range(n):
        # Calcular las predicciones
        predictions = model(X, theta_0, theta_1)
        
        # Calcular los gradientes
        gradient_theta_0 = (1 / m) * np.sum(predictions - y)
        gradient_theta_1 = (1 / m) * np.sum((predictions - y) * X)
        
        # Actualizar los parámetros
        next_theta_0 = theta_0 - learning_rate * gradient_theta_0
        next_theta_1 = theta_1 - learning_rate * gradient_theta_1
        
        # Calcular el nuevo coste
        cost = cost_function(X, y, next_theta_0, next_theta_1)
        costs.append(cost)
        
        # Comprobar si el cambio es menor que epsilon
        if abs(cost - cost_function(X, y, theta_0, theta_1)) < epsilon:
            break
        
        # Actualizar los valores de theta
        theta_0, theta_1 = next_theta_0, next_theta_1
    
    return theta_0, theta_1, costs

learning_rate = 0.1
theta_0_final, theta_1_final, cost_history = gradient_descent(X, y, theta_0_init, theta_1_init, learning_rate)
theta_0_final, theta_1_final, cost_history[-1]
````
