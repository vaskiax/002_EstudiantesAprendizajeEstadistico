# Regresión lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$


|Training|Y      | X_1  |
|--------|----------|----------|
|1|$Y^{1}$ | $X_1^{1}$|
|2|$Y^{2}$ | $X_1^{2}$|
|.|.         | .        |
|.|.         | .        |
|.|.         | .        |
|m|$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ parámetros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan más "cercano" a $Y$ para cada $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, así:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la función de coste, que permite obtener los valores más cercanos,  esta minimización podrá ser realizada a través de diferentes métodos el más conocido es el gradiente descendente.

![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)




Supongamos un modelo lineal para realizar la predicción, así nuestro modelo estará basado en la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$


Reemplando el modelo anterior en la función de coste, tenemos que:



\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

# Gradiente Descendente


Para determinar el mínimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un número aleatorios inicial $\omega_i$
- Para descender al mínimo de la función, encontremos un  valor para el cual
el deriva de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el mínimo de la función

# Modelo de *machine learning*: Solución general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

````python
def gradiente_descendente(alpha, omegai, epsilon=1e-4, max_iter=1000):
  OM = []  # Lista para almacenar los valores de x en cada iteración
  for i in range(max_iter):
     omega = gradD(omegai, alpha, fp)  # Actualizamos omega
     OM.append(omega)  # Guardamos el nuevo valor de omega (x)

        # Verificamos si el gradiente es suficientemente pequeño y paramos
     if np.abs(fp(omega)) < epsilon:
      print(f'El valor mínimo es {omega:.4f} después de {i+1} iteraciones')
      break

     omegai = omega  # Actualizamos el valor de omega para la siguiente iteración

  return OM
````

# 7. Realice una gráfica de la función en 3D, y un mapa de contorno de la función. Determine el valor mínimo de la función con el metodo del gradiente descendente.

````python
#Función dada:
def F(x, y):
  return np.sin((1/2) * x**2 - (1/4) * y**2 + 3) * np.cos(2 * x + 1 - np.exp(y))
````

````python
# Derivadas parciales:
def grad_F(x, y):
  dF_dx = np.cos((1/2) * x**2 - (1/4) * y**2 + 3) * np.cos(2 * x + 1 - np.exp(y)) * x \
            - 2 * np.sin((1/2) * x**2 - (1/4) * y**2 + 3) * np.sin(2 * x + 1 - np.exp(y))
  dF_dy = -np.sin((1/2) * x**2 - (1/4) * y**2 + 3) * np.sin(2 * x + 1 - np.exp(y)) * (1/2) * y \
            + np.cos((1/2) * x**2 - (1/4) * y**2 + 3) * np.sin(2 * x + 1 - np.exp(y)) * np.exp(y)
  return dF_dx, dF_dy
````

````python
# Gradiente descendiente:
def gradiente_descendente_2D(grad_F, x_inicial, y_inicial, alpha, epsilon=1e-4, max_iter=1000):
  x, y = x_inicial, y_inicial
  historial_x, historial_y = [x], [y]

  for i in range(max_iter):
    dF_dx, dF_dy = grad_F(x, y)
    x -= alpha * dF_dx
    y -= alpha * dF_dy

    historial_x.append(x)
    historial_y.append(y)

    if np.sqrt(dF_dx**2 + dF_dy**2) < epsilon:
      print(f"Convergencia alcanzada después de {i+1} iteraciones")
      break

  return x, y, historial_x, historial_y
````

````python
# Gráfico de la función en 3D:
def plot_3D_F():
  x = np.linspace(-2, 2, 400)
  y = np.linspace(-2, 2, 400)
  X, Y = np.meshgrid(x, y)
  Z = F(X, Y)

  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')
  ax.plot_surface(X, Y, Z, cmap='viridis')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel('F(x, y)')
  ax.set_title('Superficie de la función F(x, y)')
  plt.show()
````

````python
# Gráfico de la función en 3D:
def plot_contour_F():
  x = np.linspace(-2, 2, 400)
  y = np.linspace(-2, 2, 400)
  X, Y = np.meshgrid(x, y)
  Z = F(X, Y)

  plt.contourf(X, Y, Z, cmap='viridis', levels=50)
  plt.colorbar(label='F(x, y)')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Mapa de contorno de F(x, y)')
  plt.show()
````

````python
x_inicial, y_inicial = 0, 0
alpha = 0.01

x_min, y_min, historial_x, historial_y = gradiente_descendente_2D(grad_F, x_inicial, y_inicial, alpha)

print(f"El valor mínimo de F(x, y) se alcanza en x = {x_min:.4f}, y = {y_min:.4f}")

plot_3D_F()
````

# 9. Compare su resultado empleando la libreria linearRegresion() de sklearn.

````python
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2*X + 0.02*np.random.random(100)
````

````python
def H(theta0, theta1, X):
  h = theta0 + theta1*X
  return h
def costo(h, Y):
  return np.mean((h-Y)**2)
````

````python
def gradiente_descendente(X, y, alpha=0.01, epsilon=1e-4, max_iter=1000):
  theta0, theta1 = 0, 0  # Inicializamos los parámetros
  m = len(y)
  cost_history = []
  for i in range(max_iter):
    h = H(theta0, theta1, X)  # Calculamos la predicción
    d_theta0 = np.mean(h - y)  # Derivada de theta0
    d_theta1 = np.mean((h - y) * X)  # Derivada de theta1
    theta0 -= alpha * d_theta0
    theta1 -= alpha * d_theta1
    cost = costo(h, y)
    cost_history.append(cost)
    if len(cost_history) > 1 and np.abs(cost_history[-2] - cost_history[-1]) < epsilon:
      print(f"Convergencia alcanzada después de {i+1} iteraciones")
      break

  return theta0, theta1, cost_history
````

````python
alpha = 0.1
theta0_gd, theta1_gd, cost_history = gradiente_descendente(X, y, alpha)
print(f"Resultado del gradiente descendente: theta0 = {theta0_gd:.4f}, theta1 = {theta1_gd:.4f}")
````

````python
X_reshaped = X.reshape(-1, 1)  # Redimensionamos para sklearn
modelo = LinearRegression()
modelo.fit(X_reshaped, y)

theta0_sklearn = modelo.intercept_
theta1_sklearn = modelo.coef_[0]

print(f"Resultado de sklearn: theta0 = {theta0_sklearn:.4f}, theta1 = {theta1_sklearn:.4f}")
````

````python
plt.scatter(X, y, label="Datos reales", color = 'black')
plt.plot(X, H(theta0_gd, theta1_gd, X), color='purple', label="Gradiente descendente")
plt.plot(X, modelo.predict(X_reshaped), color='red', label="LinearRegression (sklearn)")
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Comparación: Gradiente Descendente vs LinearRegression de sklearn')
plt.grid()
plt.show()
````
